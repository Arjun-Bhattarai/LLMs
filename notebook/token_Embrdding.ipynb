A token embedding is a way to convert words (or tokens) into numbers that an AI can understand.

```python
# torch import ra input_ids tensor initialize gareko
import torch

input_ids = torch.tensor([2, 3, 5, 1])
```

```python
# vocabulary ma total kati unique tokens xan
vocab_size = 6

# harek token ko embedding vector ko size
output_dim = 3

# same random values pauna seed set gareko
torch.manual_seed(123)

# embedding layer create gareko
embedding_layer = torch.nn.Embedding(vocab_size, output_dim)
```

```python
print(embedding_layer.weight)

# yesko output ma 4 rows xa kinake input_ids ma 4 tokens set xa ra 3 columns hunxa, kina bhane output_dim 3 set gareko xain
```

```python
print(embedding_layer(torch.tensor([3])))
# tesla third row ko embedding vector output ma aauxa kina bhane input_ids ma 3 set gareko xa
```
