```python
import tiktoken
from torch.utils.data import DataLoader


def create_data_loader(
    txt,
    batch_size=5,
    max_length=256,
    stride=128,
    shuffle=True,
    drop_last=True,
    num_workers=0,
):

    # GPT tokenizer load garne
    tokenizer = tiktoken.get_encoding("gpt2")

    # dataset banaune
    dataset = GPTDataset(txt, tokenizer, max_length, stride)

    # DataLoader le batch ma data dincha
    dataloader = DataLoader(
        dataset,
        batch_size=batch_size,  # kati sample ek choti
        shuffle=shuffle,  # data random garne ki nagarne
        drop_last=drop_last,  # last incomplete batch drop garne ki nagarne
        num_workers=num_workers,  # multiprocessing workers
    )

    return dataloader
```

```python
import torch
from torch.utils.data import Dataset


class GPTDataset(Dataset):
    def __init__(self, txt, tokenizer, max_length, stride):
        self.input_ids = []
        self.target_ids = []

        token_ids = tokenizer.encode(txt)

        for i in range(0, len(token_ids) - max_length, stride):
            input_chunk = token_ids[i : i + max_length]
            target_chunk = token_ids[i + 1 : i + max_length + 1]

            # MUST be tensor
            self.input_ids.append(torch.tensor(input_chunk))
            self.target_ids.append(torch.tensor(target_chunk))

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        return self.input_ids[idx], self.target_ids[idx]
```

```python
dataloader = create_data_loader(
    txt=raw_data, batch_size=8, max_length=4, stride=4, shuffle=True
)
data_iter = iter(dataloader)
inputs, targets = next(iter(dataloader))

print(type(inputs))
print(inputs.shape)
```
