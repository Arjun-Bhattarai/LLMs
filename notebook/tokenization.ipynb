```python
# file open garera text read garna ko lagi
with open(
    r"C:\Users\Acer\Documents\LLMs\LLMs\data\the-verdict.txt", encoding="utf-8"
) as f:
    raw_data = f.read()

# total character count dekhauna
print("total characters:", len(raw_data))

# first 100 characters print garna
print(raw_data[:100])
```

```python
import re

preprocess = re.split(r'([,.:;!?()\-\[\]"\']+)', raw_data)
preprocess = [item for item in preprocess if item]
print(preprocess[:30])
```
