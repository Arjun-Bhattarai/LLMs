```python
# text lai encode and decode gerney
import re


class Tokenizer:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {i: s for s, i in vocab.items()}

    def encode(self, text):
        text = text.lower()
        text = re.sub(r"[^\w\s]", "", text)
        tokens = text.split()

        return [
            (
                self.str_to_int[token]
                if token in self.str_to_int
                else self.str_to_int["<UNK>"]
            )
            for token in tokens
        ]

    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        return text
```

```python
import tiktoken  # tiktoken library import garne (GPT style tokenizer use garna)

Tokenizer = tiktoken.get_encoding("gpt2")
# yo line le GPT-2 model le use garne BPE tokenizer load garcha
# jasma text lai subwords/tokens ma break garera encode garincha
```

```python
import tiktoken

text = (
    "Artificial intelligence has transformed modern <|endoftext|> software development, enabling programmers to automate repetitive workflows and analyze enormous datasets efficiently. However, the pseudocryptographic hypertransmogrification engine failed unexpectedly during initialization because the microfluxcapacitor produced unstable quantum-like oscillations. Researchers documented the anomaly carefully before shutting down the entire facility. "
    "Meanwhile, interdisciplinary collaboration between neuroscientists, linguists, and roboticists continues to accelerate innovation across multiple industries. The integration of AI with neuroscience has led to breakthroughs in understanding brain function, while linguistic insights have improved natural language processing algorithms. Roboticists are leveraging these advancements to create more sophisticated and adaptable machines, driving progress in fields such as healthcare, manufacturing, and autonomous systems. As these disciplines converge, the potential for transformative technologies that enhance human capabilities and address complex challenges grows exponentially."
)

integers = Tokenizer.encode(
    text,
    allowed_special={"<|endoftext|>"},
)  # yo special token lai allow garna, natra error auna sakcha


print(integers)  # text lai token ID haru ma convert bhayeko output print garne
```

```python
strings = Tokenizer.decode(integers)  # token ID haru lai text ma convert garne
print(strings)
```
