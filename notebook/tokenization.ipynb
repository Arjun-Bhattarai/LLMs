```python
# file open garera text read garna ko lagi
with open(
    r"C:\Users\Acer\Documents\LLMs\LLMs\data\the-verdict.txt", encoding="utf-8"
) as f:
    raw_data = f.read()

# total character count dekhauna
print("total characters:", len(raw_data))

# first 100 characters print garna
print(raw_data[:100])
```

```python
# text preprocessing garna ko lagi regex use garera punctuation haru split garna
import re

preprocess = re.split(r'([,.:;!?()\-\[\]"\'/@#$%^&*_+=<>|{}~`\\\n\t]+)', raw_data)
preprocess = [item for item in preprocess if item]
print(preprocess[:30])
```

```python
# unique words ko list banauna lai
text_data = raw_data.lower()
text_data = re.sub(r"[^\w\s]", "", text_data)
preprocess = text_data.split()

all_words = sorted(set(preprocess))

vocab = {word: i for i, word in enumerate(all_words)}
vocab["<UNK>"] = len(vocab)

print("vocab size:", len(vocab))
```

```python
# word lai integer ma convert garne and short ma print garne
vocab = {token: i for i, token in enumerate(all_words)}

# unknown token add garna
vocab["<UNK>"] = len(vocab)
# vocab size update garna
print("vocab size:", len(vocab))
print("sample vocab items:", list(vocab.items())[:10])
```

```python
# vocab ko first 50 items print garna
for i, item in enumerate(vocab.items()):
    print(item)

    if i > 50:
        break
```

```python
class Tokenizer:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {i: s for s, i in vocab.items()}

    # token lai integer ma convert garne
    def encode(self, text):
        text = text.lower()
        text = re.sub(r"[^\w\s]", "", text)
        tokens = text.split()

        return [
            (
                self.str_to_int[token]
                if token in self.str_to_int
                else self.str_to_int["<UNK>"]
                # <unk> la unknown token lai handle garxa
            )
            for token in tokens
        ]

    # id lai text ma convert garne
    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        return text
```

```python
# token lai integer ma convert garne tokenizer class banauna
tokenizer = Tokenizer(vocab)

text = "I felt to what a degree he had the same quality as his pictures."
ids = tokenizer.encode(text)
print("encoded ids:", ids)
```

```python
tokenizer.decode(ids)
```
